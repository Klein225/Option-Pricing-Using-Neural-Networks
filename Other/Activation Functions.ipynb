# Activation functions and their corresponding derivatives

In artificial neural networks, one of the most important concepts is the activation function. The activation function of a node defines the output of that node given it's inputs. Its main feature is enabling the neural network to capture non-linearities in the data which is crucial due to the limited nature of a linear function.


Imagine having only the identity ($f(z)=z$) as an activation function, then Equation (\ref{eq:simple_NN7}) is simply the composition of linear functions which can be replaced by a single layer (a combination of linear functions in a linear manner is still just a linear function). The NN can essentially be boiled down to simple linear regression. Also, the derivative of the identity is a constant which will cause problems when doing gradient descent as the gradient will have no relationship with $z$. It is a constant gradient and the descent is going to be on a constant gradient. if there is an error in the prediction, the changes made during back propagation is constant and will not depend on the change in the input $\Delta z$.


The optimal activation function is often a simple function as it needs to be computationally efficient. It should be chosen based on the structure of the model, the hyper-parameters and the desired output. There is no definite rule for choosing an exact activation function, so the best approach is often trial and error. If one is faced with a binary-classification problem, the best activation function would be the a soft max or binary function (binary step etc.) as a neuron should only be activated if it is above a given threshold and deactivated otherwise.


The most widely used activation functions within regression includes Sigmoid, Tanh and ReLU.

```python
import math
import pandas as pd
import datetime, os
import numpy as np
import numpy.random as npr
from pylab import plt, mpl

from scipy.stats import norm
from scipy import optimize
import scipy.integrate as integrate
import scipy.special as special 

import matplotlib.pyplot as plt
import seaborn as sns
```

Sigmoid Activation Function

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

Tanh Activation Function

```python
def tanh(z):
    return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))
```

ReLU Activation Function

```python
def relu(z):
    return np.maximum(0, z)
```

Their derivatives

```python
def derivatives(f, z, eps=0.000001):
    return (f(z + eps) - f(z - eps)) / (2 * eps)
```

Plot the activation functions and their corresponding derivatives

```python
z = np.linspace(-5, 5, 200)

plt.figure(figsize=(11,4))

plt.subplot(121)
plt.plot(z, sigmoid(z), "r--", linewidth=1.5, label="Sigmoid")
plt.plot(z, tanh(z), "g-", linewidth=1.5, label="Tanh")
plt.plot(z, relu(z), "b-.", linewidth=1.5, label="ReLU")
plt.grid(True)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=3)
plt.title("Activation functions", fontsize=12)
plt.axis([-5, 5, -1.2, 1.2])


plt.subplot(122)
plt.plot(z, derivatives(sigmoid, z), "r--", linewidth=1.5, label="Sigmoid")
plt.plot(z, derivatives(tanh, z), "g-", linewidth=1.5, label="Tanh")
plt.plot(z, derivatives(relu, z), "b-.", linewidth=1.5, label="ReLU")
plt.grid(True)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=3)
plt.title("Derivatives", fontsize=12)
plt.axis([-5, 5, -0.2, 1.2])

plt.savefig('/Users/Marcklein/Desktop/Master Thesis/Option pricing using Neural Networks/Python/Other/activation_functions.png',
            bbox_inches='tight');
plt.show()
```
